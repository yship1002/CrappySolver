```python
import json
import numpy as np
import matplotlib.pyplot as plt
with open('process.json', 'r') as f:
    crappy_data = json.load(f)
with open('/Users/jyang872/Desktop/fuzzy-tribble/process.json', 'r') as f:
    fuzzy_data = json.load(f)
fuzzy_lbd_cals=[record for record in fuzzy_data["outsideAlgo"]["LBD_calculation_records"]]
crappy_lbd_cals=crappy_data["Algo"]["lbd_calculation_record"]
```

```python
len(crappy_lbd_cals), len(fuzzy_lbd_cals)
```

```python
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.hist(fuzzy_lbd_cals, bins=30, alpha=0.7, label='FuzzySolver', edgecolor='black')
plt.xlabel('LBD Value')
plt.ylabel('Frequency')
plt.title('Frequency Distribution - FuzzySolver')
plt.legend()

plt.subplot(1, 2, 2)
plt.hist(crappy_lbd_cals, bins=30, alpha=0.7, label='CrappySolver', edgecolor='black')
plt.xlabel('LBD Value')
plt.ylabel('Frequency')
plt.title('Frequency Distribution - CrappySolver')
plt.legend()

plt.tight_layout()
plt.show()
```
